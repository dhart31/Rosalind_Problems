```python
import itertools
```

```python
length = 5
l=list(itertools.permutations([i for i in range(1,length+1)]))
s = str(len(l)) + '\n'
for m in l:

    for n in m:
        s += str(n) + ' '
    s += '\n'
```

```python
f.write(s)
f.close()
```
