```python
import numpy as np
for i in range(len(seq_list)):
    x = seq_list[i]
    v = seq_list[:i]+seq_list[i+1:]
    adj_idx=[(x.seq[-3:] == y.seq[:3]) for y in v]
    for n in range(len(adj_idx)):
        if adj_idx[n]:
            print(x.name+' '+v[n].name)
    
```
