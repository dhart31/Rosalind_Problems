```python
import numpy as np
import copy
import itertools
def count_bp(l):
    return sum(abs(np.diff(l))!=1)

def find_bp(l):
    bp_idxs = np.where(abs(np.diff(l))!=1)[0]+1
    return bp_idxs

def rev_list(l,i,j):
    t = copy.deepcopy(l)
    t[i:j] = t[i:j][::-1]
    return t

def get_strips(l,bp_idxs):
    bp_pairs = np.array(list(itertools.combinations(bp_idxs,2)))
    dec_idxs = np.array([np.any(np.diff((l[i[0]:i[1]]))==-1) for i in bp_pairs])
    return bp_pairs[dec_idxs], bp_pairs[~dec_idxs]

def reversal(l,count=0):
    global minCount
    print(l)
    if count_bp(l)==0:
        if minCount == 0 or count <= minCount:
            print('\n')
            minCount = count
    else:
        dec,inc = get_strips(l,find_bp(l))
        if len(dec)!=0:
            counts = np.zeros(len(dec))
            
            for idx,i in enumerate(dec):
                temp = rev_list(l,i[0],i[1])
                counts[idx] = count_bp(temp)
            min_idx = np.where(counts==min(counts))
            
            for i in dec[min_idx]:
                reversal(rev_list(l,i[0],i[1]),count+1)
        else:
            inc = inc[np.where(np.array([i[1]-i[0] for i in inc])!=1)]
            counts = np.zeros(len(inc))
            for idx,i in enumerate(inc):
                temp = rev_list(l,i[0],i[1])
                counts[idx] = count_bp(temp)
            print(counts)
            min_idx = np.where(counts==min(counts))
            
            for i in inc[min_idx]:
                reversal(rev_list(l,i[0],i[1]),count+1)
```
