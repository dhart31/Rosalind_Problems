```python
from Bio import SeqIO
#s=list(SeqIO.parse(open(r"C:\Users\Derek Hart\Documents\Python\rosalind_kmp.txt"),'fasta'))[0].seq
#s =list(SeqIO.read(r"C:\Users\Derek Hart\Documents\Python\rosalind_kmp.txt",'fasta').seq)
s =  list('CAGCATGGTATCACAGCAGAG')
fail = [0]*len(s)
j = 0

#iterate through each character, skipping the first
for i in range(1,len(s)):
    #if streak length is non-zero and new characters dont match
    while j>0 and s[i] != s[j]:
        j = fail[j-1] #streak of length j guaranteed to match fail[j-1] chars
    if s[i] == s[j]:
        j+=1
    fail[i] = j
    
```

```python
s =list(SeqIO.read(r"C:\Users\Derek Hart\Documents\Python\rosalind_kmp.txt",'fasta').seq)
#s = list('CAGCATGGTATCACAGCAGAG')

F_array = [0] * len(s)
k = 0
for i in range(2, len(s) + 1):
    while k > 0 and s[k] != s[i - 1]:
        k = F_array[k - 1]
    if s[k] == s[i - 1]:
        k += 1
    F_array[i - 1] = k
    
with open('array.txt', 'w') as answer:
    answer.write(' '.join(map(str, F_array)))
```
