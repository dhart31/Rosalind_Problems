```python
from Bio import SeqIO
```

```python
seqs = list(SeqIO.parse(r"C:\Users\Derek Hart\Documents\Python\rosalind_pdst.txt",'fasta'))
```

```python
slen = len(seqs[0])
for seq1 in seqs:
    row = ''
    for seq2 in seqs:
        row+=str(sum([i!=j for i,j in zip(seq1,seq2)])/slen)+' '
    print(row)
```
