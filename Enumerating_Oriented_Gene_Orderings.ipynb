```python
import itertools
n = 4
sign_perm=list(itertools.product([1,-1],repeat=n))
int_perm = list(itertools.permutations(list(range(1,n+1))))
print(len(sign_perm)*len(int_perm))
for i in range(len(sign_perm)):
    for j in range(len(int_perm)):
        x = [a*b for a,b in zip(int_perm[j],sign_perm[i])]
        print(' '.join(map(str,x)))
```
