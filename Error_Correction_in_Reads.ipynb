```python
from Bio import SeqIO
from Bio.Seq import Seq
```

```python
seqs = list(SeqIO.parse(r"C:\Users\Derek Hart\Documents\Python\rosalind_corr.txt",'fasta'))
```

```python
s=[str(i.seq) for i in seqs]
s_rc = [str(i.reverse_complement().seq) for i in seqs]
matches = list(set([i for i in s if i in s_rc]))
matches_rc=[str(Seq(i).reverse_complement()) for i in matches]
matches2 = list(set([x for x in s if s.count(x) > 1]))
matches2_rc = [str(Seq(i).reverse_complement()) for i in matches2]

all_matches = list(set(matches+matches_rc+matches2+matches2_rc))
```

```python
mismatches = [i for i in s if i not in all_matches]
```

```python
for m in mismatches:
    for n in all_matches:
        if sum([i!=j for i,j in zip(m,n)])==1:
            print(m+'->'+n)
```
