```python
num = 10

data = open(r"C:\Users\Derek Hart\Documents\Python\rosalind_tree.txt").read()
data=data.split('\n')
```

```python
n = int(data[0])
pairs = [[int(i) for i in j.split()] for j in data[1:-1]]
elements = list(range(1,n+1))
elements=[[j] for j in elements]
for pair in pairs:
    idx = [pair[0] in i for i in elements].index(True)
    idx2 = [pair[1] in j for j in elements].index(True) 
    elements[idx]+=elements[idx2]
    elements.pop(idx2)
```
